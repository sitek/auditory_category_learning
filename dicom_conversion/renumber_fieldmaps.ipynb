```python
import os
from glob import glob
```

```python
project_dir = os.path.join('/bgfs/bchandrasekaran/krs228/data/', 'FLT/')
bidsroot = os.path.join(project_dir,'data_bids')
```

```python
for subdir in sorted(glob(bidsroot+'/sub-FLT30')):
    fmapdir = subdir+'/fmap/'
    print('field map directory: ', fmapdir)

    for pe in ['AP', 'PA']: # for each phase-encoding direction (collected as a pair, but heudiconv's numberings don't align)
        json_flist = sorted(glob(fmapdir+'/*dir-{}*.json'.format(pe)))
        fmap_flist = sorted(glob(fmapdir+'/*dir-{}*.nii.gz'.format(pe)))
        assert(len(json_flist)==len(fmap_flist))
        
        print('\njson files:')
        # rename .json sidecar files
        for fx, raw_fpath in enumerate(json_flist):
            print('raw fpath: ', raw_fpath)
            raw_runid = os.path.basename(raw_fpath)[30:32]
            new_runid = f'{fx+1:02d}'

            new_fpath = raw_fpath.replace(raw_runid, new_runid)
            print('new fpath: ', new_fpath)
            print('\n')
            
            # uncomment when you're sure that it's working correctly!
            os.replace(raw_fpath, new_fpath)
        
        print('\nfmap files:')
        # rename .nii.gz fieldmap files
        for fx, raw_fpath in enumerate(fmap_flist):
            print('raw fpath: ', raw_fpath)
            raw_runid = os.path.basename(raw_fpath)[30:32]
            new_runid = f'{fx+1:02d}'

            new_fpath = raw_fpath.replace(raw_runid, new_runid)
            print('new fpath: ', new_fpath)
            print('\n')
            
            # uncomment when you're sure that it's working correctly!
            os.replace(raw_fpath, new_fpath)
```
